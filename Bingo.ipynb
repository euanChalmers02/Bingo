```python
import random
import uuid
import json

game_types = ['one_row','two_rows','bottom_top','full_house','middle']

# each has 4 random and 5 numbers
def col(start,end):
    col_num = list(range(start,end+1))

    for i in range(9):
        col_num.append('X')

    random.shuffle(col_num)
    return col_num

def get_random_id(size):
    unquie_id = str(uuid.uuid4().hex)
    return unquie_id[:size]

def add(game):
    all_cards.append(game)

def display_card(game):
    table_master = """"""
    
    raw_game = (game['game'])
    for elt in raw_game:
        table = ''
        for each in elt:
            if each == 'X':
                table = table + ' |  '+ str(each)
            elif each <=9:
                table = table + ' |  '+ str(each)
            else:
                table = table + ' | '+ str(each)
        
        row_rr = len(table)
        table_master = table_master + table
        table_master = table_master + ' | \n'
        table_master = table_master + ' '+'-'*row_rr + '\n'
    
    print("----------")           
    print('code ',game['code'])
    print("----------")
    print(table_master)

def create_cards(number):
    all_cards = []
    
    for y in range(number):
         all_cards.append(card(6))
            
    name = get_random_id(10)     
    JSON = json.dumps(all_cards)
    
    f = open(name+".json", "w")
    f.write(JSON)
    f.close()
    
    return name+".json"

def play_game(game_type,file):
    all_cards = json.loads(file)
    
    bingo_balls = []
    bingo = False
    
    while bingo == False or len(bingo_balls) <= 90:
        x = input('Action')
        if x == 'B':
            ball_temp = ball()
            bingo_balls.append(ball_temp)
            
        elif x == 'V':
            code = input('Game_Code')
            bingo = verify(code,game_type,bingo_balls)
            
def ball():
    num = random.randint(1,90)
    print(num)
    return num 

def check_card(filtered,game_type,bingo_balls):
    rez = []
    
    rr = filtered[0]
    raw_game = (rr['game'])
    for elt in raw_game:
        flag = True
        for each in elt:
            if each != 'X':
                if each not in bingo_balls:
                    flag = False
                    
        if flag == False:
            rez.append(False)
        else:
            rez.append(True)
    print(rez)
    
    if game_type == 'one_row':
        if True in rez:
            print('bingo')
    elif game_type == 'two_rows':
        if rez[0] == True and rez[1] == True:
            print('bingo')
        elif rez[1] == True and rez[2] == True:
            print('bingo')
        elif rez[0] == True and rez[2] == True:
            print('bingo')
        
    elif game_type == 'bottom_top':
        if rez[0] == True and rez[2] == True:
            print('bingo')
        
    elif game_type == 'full_house':
        if False in rez:
            print('bingo')
            
    elif game_type == 'middle':
        if rez[1] == True:
            print('bingo')
               
def verify(game_code,game_type,bingo_balls):
    filtered = [d for d in all_cards if d['code'] == game_code]
    print(filtered)
    display_card(filtered)
    
    return check_card(filtered,game_type,bingo_balls)

def card(num_of_games):
    col1 = col(1,9)
    col2 =col(10,19)
    col3 =col(20,29)
    col4 =col(30,39)
    col5 =col(40,49)
    col6 =col(50,59)
    col7 =col(60,69)
    col8 =col(70,79)
    col9 =col(80,90)
    
    card = []
    
    counter = 0
    for y in range(num_of_games):
        game = []
        for count in range(counter,counter+3):
            row =[]
            row.append(col1[count])
            row.append(col2[count])
            row.append(col3[count])
            row.append(col4[count])
            row.append(col5[count])
            row.append(col6[count])
            row.append(col7[count])
            row.append(col8[count])
            row.append(col9[count])
            game.append(row)
        
        counter = counter + 3
        dict_game = {'game':game, 'code':'T'+ get_random_id(4)}
        card.append(dict_game)
        add(dict_game)
        
    return card
```
